```python
from Fake_news_detection import manual_testing

def fakenewspredict(t):
     return manual_testing(t)
```

```python
from text_summarization import summarization

def txtsum(t):
     return summarization(t)
```

```python
from codegen import gencode

def codegen(t):
     return gencode(t)
```

```python
from machine_translation import translate

def predictlanguage(t):
     return translate(t)
```

```python
import tkinter as tk
from tkinter import ttk

def button_clicked(input_field, text_view,func):
    text = input_field.get("1.0", tk.END).strip()  # Retrieve the text from the text area
    text=func(text)
    text_view.config(state=tk.NORMAL)
    text_view.delete("1.0", tk.END)
    text_view.insert(tk.END, text)
    text_view.config(state=tk.DISABLED)

root = tk.Tk()
root.title("NLP Project 2023")

tab_control = ttk.Notebook(root)

# Create the first tab
tab1 = ttk.Frame(tab_control)
tab_control.add(tab1, text="Tab 1")

header1 = ttk.Label(tab1, text="Fake News", font=("TkDefaultFont", 16, "bold"))
header1.pack()

input_field1 = tk.Text(tab1, width=70, height=10)
input_field1.pack()

button1 = ttk.Button(tab1, text="classify news", command=lambda: button_clicked(input_field1, text_view1,fakenewspredict), width=20)
button1.pack()

text_view1 = tk.Text(tab1, state=tk.DISABLED)
text_view1.pack()

# Create the second tab
tab2 = ttk.Frame(tab_control)
tab_control.add(tab2, text="Tab 2")

header2 = ttk.Label(tab2, text="Text Summarization", font=("TkDefaultFont", 16, "bold"))
header2.pack()

input_field2 = tk.Text(tab2, width=70, height=10) 
input_field2.pack()

button2 = ttk.Button(tab2, text="Summarize", command=lambda: button_clicked(input_field2, text_view2,txtsum), width=20)
button2.pack()

text_view2 = tk.Text(tab2, state=tk.DISABLED)
text_view2.pack()

# Create the third tab
tab3 = ttk.Frame(tab_control)
tab_control.add(tab3, text="Tab 3")

header3 = ttk.Label(tab3, text="code generation", font=("TkDefaultFont", 16, "bold"))
header3.pack()

input_field3 = tk.Text(tab3, width=70, height=10) 
input_field3.pack()

button3 = ttk.Button(tab3, text="generate", command=lambda: button_clicked(input_field3, text_view3,codegen), width=20)
button3.pack()

text_view3 = tk.Text(tab3, state=tk.DISABLED)
text_view3.pack()



# Create the third tab
tab4 = ttk.Frame(tab_control)
tab_control.add(tab4, text="Tab 4")

header4 = ttk.Label(tab4, text="Machine translation", font=("TkDefaultFont", 16, "bold"))
header4.pack()

input_field4 = tk.Text(tab4, width=70, height=10) 
input_field4.pack()

button4 = ttk.Button(tab4, text="translate", command=lambda: button_clicked(input_field4, text_view4,predictlanguage), width=20)
button4.pack()

text_view4 = tk.Text(tab4, state=tk.DISABLED)
text_view4.pack()



tab_control.pack(expand=1, fill="both")

root.mainloop()
```
